```python
from os import listdir
from PIL import Image
import os.path
import numpy as np

path_new = 'C:/Users/Admin/Downloads/HMM/Proper/Exe/Experiment 1/Train 2'
h = 32 #height of image
w = 32 #width of image

#be careful with using this function, it will consume memory, access to disk and time
timid = []
for f in listdir(path_new):
    with open(os.path.join(path_new, f), 'rb') as img_set:
        img_arr = img_set.read(h*w)
        while img_arr:
            if len(img_arr) == h*w and img_arr not in timid:
                timid.append(img_arr)
            img_arr = img_set.read(h*w)

            
#And you can save them into png files
count = 1
for img in timid:
    png = Image.fromarray(np.reshape(list(img), (h,w)).astype('float32'), mode='L')
    png.save('Exe/Experiment 1/Train 2 Img/%d.png'%count)
    count += 1
```

```python
from os import listdir
from PIL import Image
import os.path
import numpy as np

path_new = 'C:/Users/Admin/Downloads/HMM/Proper/Exe/Experiment 2/Train2'
h = 32 #height of image
w = 32 #width of image

#be careful with using this function, it will consume memory, access to disk and time
intruder = []
for f in listdir(path_new):
    with open(os.path.join(path_new, f), 'rb') as img_set:
        img_arr = img_set.read(h*w)
        while img_arr:
            if len(img_arr) == h*w and img_arr not in intruder:
                intruder.append(img_arr)
            img_arr = img_set.read(h*w)

            
#And you can save them into png files
count = 1
for img in intruder:
    png = Image.fromarray(np.reshape(list(img), (h,w)).astype('float32'), mode='L')
    png.save('Exe/Experiment 2/Train 2 Img/%d.png'%count)
    count += 1
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
import numpy as np

#reshape images to fit into the CNN model
img_list = np.zeros(shape = (len(timid), h,w,1), dtype = np.uint8)
for j in range(len(timid)):
    img_list[j,:,:,0] = np.reshape(list(timid[j]), (h,w))
    
img_list = img_list.astype('float32')
img_list /= 255
```

```python
#reshape images to fit into the CNN model
img_list1 = np.zeros(shape = (len(intruder), h,w,1), dtype = np.uint8)
for j in range(len(intruder)):
    img_list1[j,:,:,0] = np.reshape(list(intruder[j]), (h,w))
    
img_list1 = img_list1.astype('float32')
img_list1 /= 255

img_list1
```

```python
x = np.array([i for i in img_list]+[i for i in img_list1])
model = Sequential()
#Conv2D Layers
model.add(Conv2D(12, (25, 25), padding='same',input_shape=(x).shape[1:], activation = 'relu'))
model.add(Conv2D(12, (25, 25), padding='same', activation = 'relu'))
#Max Pooling Layer
model.add(MaxPooling2D(pool_size=(2, 2)))
#Conv2D Layer
model.add(Conv2D(12, (13, 13), padding='same', activation = 'relu'))
model.add(Conv2D(12, (13, 13), activation = 'relu'))
#Max Pooling
model.add(MaxPooling2D(pool_size=(2, 2)))
#Flattening Layer
model.add(Flatten())
#Dense Layer
model.add(Dense(1024, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['binary_accuracy'])

model.summary()
```

```python
epochs = 100
labels = [0 for _ in img_list] + [1 for _ in img_list1]
model.fit(np.array(x), np.array(labels),
              epochs = epochs,
          validation_split = 0.25,
              shuffle = True)
```
